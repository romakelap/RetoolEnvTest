# Read Input

```python
# Create a ZipFile object
df = None
with zipfile.ZipFile(zip_path, 'r') as z:
    # List to store DataFrames
    dfs = []

    # Loop over each file in the zip file
    for filename in z.namelist():
        # Check if the file is a CSV
        try:
          if filename.endswith('.csv'):
              # Open each file
              with z.open(filename) as f:
                  # Convert bytes to text
                  data = io.TextIOWrapper(f)
                  # Read the CSV data
                  df_tmp = pd.read_csv(data, encoding= 'utf-8')
                  df_tmp['file_name'] = filename
                  # Append the DataFrame to the list
                  dfs.append(df_tmp)
        except (ParserError):
          print(filename)
          continue

    # Combine all DataFrames in the list into a single DataFrame
    df = pd.concat(dfs)

# csv_path = '/content/drive/My Drive/External/livestream_raw_20230705.csv'
# df = pd.read_csv(csv_path)

# Convert 'account_datetime' to datetime format
df['account_datetime'] = pd.to_datetime(df['account_datetime'])

# UTC+
df['account_datetime'] = df['account_datetime'] + pd.Timedelta(hours=timezone)
# UTC-
#df['account_datetime'] = df['account_datetime'] - pd.Timedelta(hours=timezone)

if is_filter_datetime:
    # Filter by month
    df = df[df['account_datetime'].dt.month == month_filter]
    df = df[df['account_datetime'].dt.year == year_filter]

# Filter by handle
df['account_handle'] = df['account_handle'].str.strip().str.lower()
df = df[df['account_handle'].isin(account_list)]
df = df.sort_values(by=['account_datetime'], ascending=True)

df.to_csv('df_raw.csv', index=False)

for account_handle in account_list:
    if account_handle not in [str(acc).strip().lower() for acc in list(df['account_handle'])]:
      print(account_handle)

df
```

# Preprocessing

```python
null_counts = df[df['account_datetime'].isnull()].groupby('file_name').size()

# Sort the counts in descending order
sorted_counts = null_counts.sort_values(ascending=False)

# Print the sorted counts
print(sorted_counts)

# Drop empty rows
df.dropna(subset=['account_datetime', 'account_handle'], inplace=True)
```

```python
# Create 'is_livestreaming' column based on 'account_room_id'
df['is_livestreaming'] = df['account_room_id'].notna()

# Truncate datetime to half-hours
df_anl = df[['account_handle', 'account_datetime', 'is_livestreaming']].copy()  # Create a copy of the DataFrame
df_anl.loc[:, 'account_datetime'] = df_anl['account_datetime'].dt.floor('30T')
# df_anl.loc[:, 'account_datetime'] = df_anl['account_datetime'].dt.floor('H')

df_anl = df_anl.drop_duplicates()
df_anl = df_anl.dropna()
df_anl.to_csv('df_anl.csv', index=False)
df_anl
```

# Liveselling Analysis v2

# [OBSOLETE] Average Daily Stream

```python
df_grouped = df_anl.copy()

# Total days
total_days = len(sorted([str(x) for x in list(set(df_grouped['account_datetime'].dt.date))]))

# Filter by active livestreams
df_grouped = df_grouped[df_grouped['is_livestreaming'] == True]
df_grouped = df_grouped.drop_duplicates()
df_grouped = df_grouped.dropna()
df_grouped.to_csv('df_grouped.csv', index=False)

# Group by
df_grouped = df_grouped.groupby(['account_handle']).size().reset_index()
df_grouped.columns = ['account_handle', 'stream_count']
# df_grouped.loc[:, 'total_stream_hours'] = df_grouped['stream_count'] * 1
df_grouped.loc[:, 'total_stream_hours'] = df_grouped['stream_count'] * 0.5
df_grouped.loc[:, 'avg_daily_stream_hours'] = df_grouped['total_stream_hours'] / total_days

# Insert missing handles
for account_handle in account_list:
    if account_handle not in list(df_grouped['account_handle']):
        df_handle = pd.DataFrame([[account_handle]], columns=['account_handle'])
        df_grouped = pd.concat([df_grouped, df_handle])

# Sort by highest
df_grouped = df_grouped.sort_values(by=['avg_daily_stream_hours'], ascending=False)

# Finalize
df_grouped = df_grouped.drop('stream_count', axis=1)
df_grouped = df_grouped.fillna(0)
df_grouped.loc[:, 'total_tracked_days'] = total_days
df_grouped.to_csv('avg_daily_stream.csv', index=False)
df_grouped
```

# [OBSOLETE] Hourly Streaming Heatmap

```python
df_hourly = df_anl.copy()

df_hourly['account_datetime'] = df_hourly['account_datetime'] + pd.Timedelta(hours=7)
df_hourly['time'] = df_hourly['account_datetime'].dt.floor('h').dt.time
df_hourly = df_hourly[df_hourly['is_livestreaming'] == True]
df_hourly = df_hourly.drop_duplicates()
df_hourly = df_hourly.dropna()

df_hourly = df_hourly.groupby('time').agg({'account_handle': 'nunique'}).reset_index()
df_hourly.columns = ['stream_hour', 'total_recorded']

df_hourly.to_csv('df_hourly.csv', index=False)
df_hourly
```

# [FINAL]

```python
df_live = df_anl.copy()
df_live['is_livestreaming'] = df_live['is_livestreaming'].apply(lambda x: 1 if x else 0)
df_live['hour'] = df_anl['account_datetime'].dt.hour
df_live.drop_duplicates(inplace=True)
df_live
```

```python
df_live_agg = df_live.groupby(['account_handle', 'hour']).agg({'is_livestreaming': 'mean'}).reset_index()
df_live_agg
```

```python
df_live_agg_acc = df_live_agg.groupby(['account_handle']).agg({'is_livestreaming': 'sum'}).reset_index()
df_live_agg_acc['percentage_of_liveselling'] = df_live_agg_acc.apply(lambda x: x['is_livestreaming']/24, axis=1)
df_live_agg_acc = df_live_agg_acc.sort_values(by=['is_livestreaming'], ascending=False)

df_live_agg_acc.to_csv('df_live_agg_acc.csv', index=False)
df_live_agg_acc
```

```python
df_live_agg_hour = df_live_agg.groupby(['hour']).agg({'is_livestreaming': 'sum'}).reset_index()
df_live_agg_hour['percentage_of_acc'] = df_live_agg_hour.apply(lambda x: x['is_livestreaming']/len(df_live_agg_acc), axis=1)

df_live_agg_hour.to_csv('df_live_agg_hour.csv', index=False)
df_live_agg_hour
```
